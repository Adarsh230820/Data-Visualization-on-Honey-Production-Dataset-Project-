```python
# importing required libraries

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# reading dataset

data_df = pd.read_csv('honeyproduction.csv')
```

```python
# 3. Display the percentage distribution of the data in each year using the pie chart.

data_df.year.value_counts()             # to get distribution of data in every year as count
```

```python
# plotted pie chart using value_counts of year column

plt.pie(data_df.year.value_counts(), autopct = '%.1f%%',labels = data_df.year.value_counts().index)
plt.title('Percentage distribution of the data in each year')
plt.show()
```

```python
# 4. Plot and Understand the distribution of the variable "price per lb" using displot, and write your findings.

# plotted displot of priceperlb column data

sns.displot(data=data_df,x='priceperlb',kde=True)
plt.title('Distribution of the \"price per lb\"')
plt.show() 
```

```python
# 5. Plot and understand the relationship between the variables 'numcol' and 'prodval' through scatterplot, and write your findings.

# plotted scatter plot of numcol vs prodvalue columns data

sns.scatterplot(data=data_df,x='numcol',y='prodvalue')
plt.title('\'numcol\' and \'prodvalue\' relationship')
plt.show()
```

```python
# 6. Plot and understand the relationship between categorical variable 'year' and a numerical variable 'prodvalue' through boxplot, and write your findings.

plt.figure(figsize=(10,7.5))
sns.boxplot(data=data_df,x='year',y='prodvalue')
plt.title('Boxplot using \'year\',\'prodvalue\'')
plt.show()
```

```python
# 7. Visualize and understand the relationship between the multiple pairs of variables throughout different years using pairplot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'prodvalue','year')

data_df1 = data_df[['numcol','yieldpercol','totalprod','prodvalue','year']]
sns.pairplot(data=data_df1,hue='year')
plt.show
```

```python
# 8. Display the correlation values using a plot and add your inferences. (use columns 'numcol', 'yield percol','total prod', 'stocks', 'price per lb', 'prodvalue')

data_df2=data_df[['numcol','yieldpercol','totalprod','stocks','priceperlb','prodvalue']]
data_df2.corr()
```

```python
sns.heatmap(data_df2.corr(),annot=True,linewidths=1)
plt.title('Correlation Matrix')
plt.show()
```
